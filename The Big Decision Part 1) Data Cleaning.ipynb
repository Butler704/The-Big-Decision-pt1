# Project The Big Decision Part 1: Data Cleaning


##  For this project your job is to clean the following dataset and put it in the best form for analysis.
###  For this project you will need to do the following:
#### - Split columns
#### - Change columns data types
#### - Change the  columns units
#### -  Replace NaN values
#### - Filter the dataset based on specific conditions


```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
nba=pd.read_csv("nba2k-full.csv") # this version will show the cleaned dataset
nba_OG=pd.read_csv("nba2k-full.csv") # this version will show the original dataset with no changes made
nba.head()
```

#### Split the Full name column into two seperate columns called "First_Name" and "Last_Name"

```python
nba[['First_Name','Last_Name']]=nba.full_name.str.split(' ',expand=True).rename(columns={0:'First_Name',1:'Last_Name'}).drop(columns=2)
first=nba.pop('First_Name')
second=nba.pop('Last_Name')
nba.insert(0,'First_Name',first)
nba.insert(1,'Last_Name',second)
nba.head()
```

#### Drop the full name column from the dataset

```python
nba.drop(columns="full_name",inplace=True)
nba
```

#### Drop the '#' in front of the Jersey number and transform the values in the jersey column into integers.

```python
nba['jersey']=nba.jersey.str.split('#',expand=True).drop(columns=0).rename(columns={1:'jeresy'}).astype(int)
```

#### Break up the position column into two columns the first one being called "position" and the second being called "secondary_position"

```python
nba[["position","secondary_position"]]=nba.position.str.split('-',expand=True).rename(columns={0:"primary_position",1:"secondary_position"})
```

```python
sixth=nba.pop("secondary_position")
nba.insert(6,"secondary_position",sixth)
nba.head()
```

####  Drop the height by meters and convert height in feet and inches to just inches

```python
x=nba.height.str.split(expand=True).drop(columns=[1,2])
x=x[0].str.split("-",expand=True)
x[[0,1]]=x[[0,1]].astype(int)
x.dtypes
nba['height']=x[0]*12+x[1]
```

```python
nba.rename(columns={'height':'height(inches)'},inplace=True)
```

#### Drop the Weight in Kilograms and any other characters in the weight column

```python
nba['weight']=nba['weight'].str.split(expand=True).drop(columns=[1,2,3,4])
```

```python
nba.rename(columns={'weight':'weight(lbs)'},inplace=True)
```

#### Remove the dollar sign from salary and convert the values in this column into integers

```python
nba['salary']=nba.salary.str.strip('$').astype(int)
```

#### Change the format salary unit from dollars to by millions of dollars and rename the column 'salary_mil'

```python
nba['salary']=nba.salary/1000000
```

```python
nba['salary']=nba['salary'].round(decimals=1)
nba.rename(columns={"salary":"salary_mil"},inplace=True)
```

```python
nba[['salary_mil']].head(3)
```

```python
nba.head(3)
```

#### In the 'draft_peak' column if the value ="Undrafted"  then change it to 0.Second, change the name of  draft_peak column to draft_pick. Finally, change the data type for draft_pick to an integer.

```python
nba.draft_peak=nba.draft_peak.replace('Undrafted',0)
```

```python
nba.rename(columns={'draft_peak':'draft_pick'},inplace=True)
```

```python
#nba['draft_classification']
nba['draft_pick']=nba['draft_pick'].astype(int)
```

#### Create a new column called "lottery_picks". If a players draft pick is less than or equal to 15 than put "Lottery" else put "Not_Lottery".

```python
nba['lottery_picks']=nba['draft_pick'].apply( lambda x: "Lottery" if x>=1 and x<=15 else "Not_Lottery")
```

```python
nba['lottery_picks'].value_counts()
```

#### For the college column replace all NaN values with "Other"

```python
nba.college.value_counts(dropna=False).head()
```

```python
nba.college=nba.college.fillna("Other")
```

```python
nba.college.value_counts().head()
```

#### Filter out the other versions so that we only see players from NBA2k20

```python
nba=nba.loc[nba.version=='NBA2k20',]
```

```python
nba.version.value_counts()
```
